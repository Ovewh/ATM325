```python
import xarray as xr
import matplotlib.pyplot as plt
import cartopy as cr
import cartopy.crs as ccrs
import matplotlib as mpl
from mpl_toolkits.axes_grid1 import make_axes_locatable
from matplotlib.colors import BoundaryNorm
import numpy as np
from plotYearlyGridded import plot_yearly_gridded
```

```python
paths = ['data/NO2/OMNO2d_003_Column20050101-20051231_Global.nc', 
        'data/NO2/OMNO2d_003_Column.20190101-20191231_Global.nc']
fig, ax = plot_yearly_gridded(paths[0],vmax=2.6e16 ,title='2005 time averaged tropospheric NO2 over China', cmap='viridis')
ax.set_extent([90,140,20,45], crs=ccrs.PlateCarree())
ax.add_feature(cr.feature.BORDERS)
plt.savefig('figs/China_NO2_2005.png', dpi = 300, bbox_inches='tight')
```

```python
fig, ax = plot_yearly_gridded(paths[1],vmax=2.6e16 ,title='2019 time averaged tropospheric NO2 over China', cmap='viridis')
ax.set_extent([90,140,20,45], crs=ccrs.PlateCarree())
ax.add_feature(cr.feature.BORDERS)
plt.savefig('figs/China_NO2_2019.png', dpi = 300, bbox_inches='tight')
```
